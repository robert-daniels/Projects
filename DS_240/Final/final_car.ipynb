```python
x = df[['year', 'km_driven']]
y = df['selling_price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 50)
```

```python
lm = LinearRegression()
lm.fit(x_train, y_train)
```

```python
print("Intercept: ", lm.intercept_)
print(lm.coef_)

print("Coeffs - Year: ", lm.coef_[0])
print("Coeffs: km_driven: ", lm.coef_[1])
```
